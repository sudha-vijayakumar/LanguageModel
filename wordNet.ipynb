```python
import requests 

id="00001740-n"
response = requests.get("http://wordnet-rdf.princeton.edu/json/id/"+id)
data = response.json()
dic = data[0]
defn = dic['definition']
lemmas = dic['lemmas']
word = (lemmas[0])['lemma']
subject = dic['subject']
pos = dic['pos']
lang = (lemmas[0])['language']
print(word,':',defn,':',pos,':',subject,':',lang)
```
